```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import re
```

```python
# add values on top of each bar
def bar_annotation(ax):
    for p in ax.patches:
        ax.annotate(np.round(p.get_height(),decimals=2), (p.get_x()+p.get_width()/2., p.get_height()), \
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
```

```python
loc_case = pd.read_csv("dataset/processed_location_Sep20th2020.csv")
idiv_case = pd.read_csv("dataset/processed_individual_cases_Sep20th2020.csv")
```

# 1.2 Data cleaning and Imputing missing values

## First work on cases based on location dataset

### The number of values missing is small compared to the size of entire dataset (3954)
### Plus it is hard to meaningfully impute values for province, location and incidence rate

```python
nan_count = loc_case.isnull().sum()
ax = nan_count.plot(kind='bar', rot=45, title = "Number of Missing Values for Attributes")
ax.set_ylabel('# of missing values')
bar_annotation(ax)
```

### Investigation on Case-Fatality_Ratio shows that all missing values are due to the 0 # of confirmed cases

```python
fatality_miss = loc_case[['Confirmed', 'Case-Fatality_Ratio']]
fatality_miss = fatality_miss[fatality_miss['Case-Fatality_Ratio'].isnull()].sort_values(by = 'Confirmed', ascending = False)
print(len(fatality_miss))
print(fatality_miss[:10])
```

### Thus, we can safely remove all records with Nan in the location based dataset without losing the essence of the dataset

```python
loc_case = loc_case.dropna()
```

```python
loc_case.describe()
```

## Data cleaning and impute for individual cases dataset

```python
nan_count = idiv_case.isnull().sum()
missing_value_plot = nan_count.plot(kind='bar', rot=45, title = "Number of Missing Values for Attributes")
missing_value_plot.set_ylabel('# of missing values')
bar_annotation(missing_value_plot)
```

### Reduce different formats of age to a standard format and create new column. X~Y --> (X+Y)/2, N months --> N / 12

```python
i = 0;
def parse_age(age):
    global i
    if pd.isnull(age):
        return
    age_without_dash = age.replace("-", " ")
    age_list = re.findall(r"[+-]? *(?:\d+(?:\.\d*)?|\.\d+)(?:[eE][+-]?\d+)?", age_without_dash) # https://stackoverflow.com/questions/45001775/find-all-floats-or-ints-in-a-given-string
    #print(age_list)
    if len(age_list) > 1 and age_list[-1] == '0': ## age string that is float (20.0) will be parsed into lsit [20, 0]
        age_list.pop();

    parsed_age = int(sum(float(x) for x in age_list)/len(age_list))

    if re.search('month', age_without_dash, re.IGNORECASE): ## age string that contains month
        print(age_without_dash, 'replaced with', int(parsed_age / 12))
        return int(parsed_age / 12)

    if i < 10 and len(age_list) > 1: ## print first 10 ages with format x-y
        i = i+1
        print(age, 'replaced with', parsed_age)
    return parsed_age
```

```python
age_parsed = idiv_case['age'].apply(lambda x: parse_age(x))
idiv_case['age'] = age_parsed
```

### For missing age values, cannot meaningfully impute with mean or mode since more than half of age values are missing. 
### Set default -1 to null values

```python
idiv_case['age'] = idiv_case['age'].fillna(-1)
```

### Too many missing values in attribute additional_information (about 94% missing), remove the entire column
### Many missing values in attribute source, cannot meaningfully impute and the attribute is intuitively irrelvant. Remove column as well

```python
idiv_case = idiv_case.drop(columns = ['additional_information', 'source'])
```

### For sex attribute, cannot meaningfuly impute, fill Nan with unknown

```python
idiv_case['sex'] = idiv_case['sex'].fillna('unknown')
```

### For all other attributes, since only a small number of values are missing and it is hard to impute meaningfully for attributes like province, longitude and latitude etc, can safely drop without losing the essence of the dataset

```python
idiv_case = idiv_case.dropna()
idiv_case
```

# 1.3 Dealing with outliers

```python
def remove_outlier_by_iqr(df, column):
    first_qt = df[column].quantile(0.25)
    third_qt = df[column].quantile(0.75)
    iqr = (third_qt - first_qt)*1.5
    lower_bound = first_qt - iqr
    upper_bound = third_qt + iqr
    print('First Quantile: ', first_qt, 'Third Quantile: ', third_qt)
    print('Lower Bound: ', lower_bound, "Upper Bound: ", upper_bound)
    return df[(df[column] > upper_bound) | (df[column] < lower_bound)]


def remove_outlier_by_z(df, column, std):
    df['z_score'] = stats.zscore(df[column])
    outlier = df[np.abs(df['z_score']) > std].drop(columns = ['z_score'])
    del df['z_score']
    return outlier
```
